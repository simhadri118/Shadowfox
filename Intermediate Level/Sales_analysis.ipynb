#1. Sales Data Analysis Project

This project analyzes sales data to understand trends, profit distribution, and the impact of discounts.

#2. Import Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#3. Load Dataset

```python
data = pd.read_csv("Sales.csv")
data.head()
```

#4. Data Overview
Checking structure and summary of dataset.

```python
data.info()
data.describe()
```

#5. Data Cleaning
Handling missing values and formatting dates.

```python
data.isnull().sum()

data["Order Date"] = pd.to_datetime(data["Order Date"])
```

# **DATA VISUALIZATION**

#6. Sales by Region
This graph shows total sales across different regions.

```python
data.groupby("Region")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

#7. Sales by category
This graph shows which category generates the most revenue.

```python
data.groupby("Category")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Category")
plt.show()
```

#8. Profit by Region
Analyzing profit distribution across regions.

```python
data.groupby("Region")["Profit"].sum().plot(kind="bar")
plt.title("Profit by Region")
plt.show()
```

#9. Monthly Sales Trend
This graph shows how sales vary month-wise.

```python
data["Month"] = data["Order Date"].dt.month

data.groupby("Month")["Sales"].sum().plot(kind="line", marker='o')
plt.title("Monthly Sales Trend")
plt.show()
```

#10. Top Five Products
Identifying highest selling products.

```python
top_products = data.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(5)

top_products.plot(kind="bar")
plt.title("Top 5 Products")
plt.show()
```

#11. Discount vs Profit
Analyzing how discounts impacts profit.

```python
sns.scatterplot(x="Discount", y="Profit", data=data)
plt.title("Discount vs Profit")
plt.show()
```

#12. Sales Distribution
Understanding distribution of sales values.

```python
sns.histplot(data["Sales"])
plt.title("Sales Distribution")
plt.show()
```

#13. Correlation Heatmap
Shows relationships between numerical variables.

```python
numerical_data = data.select_dtypes(include=['number'])
sns.heatmap(numerical_data.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

#14. Pair Plot
Visual comparison between multiple variables.

```python
sns.pairplot(data)
plt.show()
```

#15. Average Sales by Region

```python
print(data.groupby("Region")["Sales"].mean())
```

#16. Total Profit

```python
print("Total Profit:", data["Profit"].sum())
```

# Final Insights

- East region has highest sales and profit
- Technology category contributes most revenue
- Sales show variation across months
- Higher discount reduces profit
- Few products contribute major revenue
